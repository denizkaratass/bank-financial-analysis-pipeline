```python
import pandas as pd
import glob
import os
import re

# 1. Verileri Oku ve Listele
path = "." # Dosyaların olduğu yer
files = glob.glob(os.path.join(path, "Kar Zarar_*.xlsx"))
data = []

for f in files:
    df = pd.read_excel(f)
    fname = os.path.basename(f)
    
    # Dosya isminden tarih bilgisini güvenli bir şekilde al
    match = re.search(r'Kar Zarar_(\d{4})_(\d{1,2})', fname)
    yil, ay = int(match.group(1)), int(match.group(2))
    tarih_str = f"{yil}-{ay:02d}-01 00:00:00"
    
    # 'DÖNEM NET KARI' satırını bul
    label_col = [c for c in df.columns if 'Kar Zarar' in str(c)][0]
    row = df[df[label_col].str.contains('DÖNEM NET KARI', na=False)]
    
    if not row.empty:
        data.append({
            'tarih': tarih_str,
            'year': yil,
            'kalem': 'DÖNEM NET KARI (ZARARI) (51-52)',
            'Toplam': row['Toplam'].values[0]
        })

# 2. DataFrame Oluştur
df_final = pd.DataFrame(data)

# 3. Tarihe Göre Sırala
df_final = df_final.sort_values(by='tarih').reset_index(drop=True)

# 4. Aylık Kârı Hesapla
# Sıralı veri üzerinden bir önceki aydan farkı alıyoruz
df_final['gercek_aylik_kar'] = df_final.groupby('year')['Toplam'].diff().fillna(df_final['Toplam'])

# 5. Kaydet
df_final.to_excel("Faz1_NetKar_Sonuc_Birebir.xlsx", index=False)
print("İşlem Tamam")
```
