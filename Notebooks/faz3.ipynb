```python
import pandas as pd
import glob
import os
import re

def faz3_olustur():
    # 1. Sadece Excel Dosyalarını Tespit Et
    path = "."
    files = glob.glob(os.path.join(path, "Kar Zarar_*.xlsx"))

    if not files:
        print("HATA: Klasörde 'Kar Zarar_' ile başlayan Excel dosyası bulunamadı!")
        return

    files.sort() # Kronolojik sıra için
    raw_list = []

    for f in files:
        df = read_func(f)
        fname = os.path.basename(f)
        
        # Tarih bilgisini dosya adından ayıkla
        match = re.search(r'Kar Zarar_(\d{4})_(\d{1,2})', fname)
        yil, ay = int(match.group(1)), int(match.group(2))
        tarih_str = f"{yil}-{ay:02d}-01 00:00:00"
        
        # Dinamik sütun ismi tespiti
        label_col = [c for c in df.columns if 'Kar Zarar' in str(c)][0]
        
        # KÜMÜLATİF (YTD) VERİLERİ ÇEK
        # Dönem Net Karı
        net_kar_kum = df[df[label_col].str.contains('DÖNEM NET KARI', na=False)]['Toplam'].values[0]
        # Diğer Riskler -> 'Diğer Provizyonlar' satırından gelir
        diger_risk_kum = df[df[label_col].str.contains('Diğer Provizyonlar', na=False)]['Toplam'].values[0]
        # Kredi Karşılıkları -> 'Genel Karşılık Provizyonu' satırından gelir
        kredi_kars_kum = df[df[label_col].str.contains('Genel Karşılık Provizyonu', na=False)]['Toplam'].values[0]
        
        raw_data.append({
            'tarih': tarih_str,
            'year': yil,
            'Kum_Net_Kar': net_kar_kum,
            'Kum_Diger_Risk': diger_risk_kum,
            'Kum_Kredi_Kars': kredi_kars_kum
        })

    # 2. DataFrame ve Aylık Fark Hesaplamaları
    df_raw = pd.DataFrame(raw_data).sort_values('tarih').reset_index(drop=True)
    
    # Birikimli verileri birbirinden çıkararak o ayın net rakamlarını buluruz
    df_raw['Net_Kar'] = df_raw.groupby('year')['Kum_Net_Kar'].diff().fillna(df_raw['Kum_Net_Kar'])
    df_raw['Diger_Riskler'] = df_raw.groupby('year')['Kum_Diger_Risk'].diff().fillna(df_raw['Kum_Diger_Risk'])
    df_raw['Kredi_Karsiliklari'] = df_raw.groupby('year')['Kum_Kredi_Kars'].diff().fillna(df_raw['Kum_Kredi_Kars'])
    
    # 3. Final Tablo Formatı 
    cols = ['tarih', 'Diger_Riskler', 'Net_Kar', 'Kredi_Karsiliklari']
    df_faz3 = df_raw[cols]
    
    # Kaydet
    df_faz3.to_excel("Faz3_Risk_Analizi.xlsx", index=False)
    print("------------------------------------------")
    print("BAŞARILI: 'Faz3_Risk_Analizi.xlsx' dosyası birebir oluşturuldu!")
    print("------------------------------------------")
    return df_faz3

if __name__ == "__main__":
    generate_faz3_birebir()
```
