```python
import pandas as pd
import glob
import os
import re

def faz2_olustur():
    # 1. Excel Dosyalarƒ±nƒ± Tespit Et
    path = "."
    files = glob.glob(os.path.join(path, "Kar Zarar_*.xlsx"))

    if not files:
        print("‚ùå HATA: Klas√∂rde 'Kar Zarar_' ile ba≈ülayan Excel dosyasƒ± bulunamadƒ±!")
        return

    files.sort() 
    raw_list = []

    print(f"üöÄ {len(files)} adet Excel dosyasƒ± Faz 2 i√ßin i≈üleniyor...")

    for f in files:
        try:
            # Excel dosyasƒ±nƒ± oku
            df = pd.read_excel(f)
            fname = os.path.basename(f)
            
            # Tarih bilgisini dosya adƒ±ndan √ßek (√ñrn: Kar Zarar_2023_01)
            match = re.search(r'Kar Zarar_(\d{4})_(\d{1,2})', fname)
            yil, ay = int(match.group(1)), int(match.group(2))
            tarih_str = f"{yil}-{ay:02d}-01 00:00:00"
            
            # Dinamik s√ºtun tespiti (Ba≈ülƒ±k ismi her ay deƒüi≈üebilir)
            label_col = [c for c in df.columns if 'Kar Zarar' in str(c)][0]
            
            # K√ºm√ºlatif verileri √ßek 
            net_kar = df[df[label_col].str.contains('D√ñNEM NET KARI', na=False)]['Toplam'].values[0]
            gelir = df[df[label_col].str.contains('Toplam Faiz \(Kar Payƒ±\) Gelirleri', na=False)]['Toplam'].values[0]
            gider = df[df[label_col].str.contains('Toplam Faiz \(Kar Payƒ±\) Giderleri', na=False)]['Toplam'].values[0]
            marj = df[df[label_col].str.contains('NET FAƒ∞Z \(KAR PAYI\) GELƒ∞Rƒ∞', na=False)]['Toplam'].values[0]
            
            raw_list.append({
                'tarih': tarih_str,
                'year': yil,
                'K_Net_Kar': net_kar,
                'K_Gelir': gelir,
                'K_Gider': gider,
                'K_Marj': marj
            })
        except Exception as e:
            print(f"‚ö†Ô∏è {fname} i≈ülenirken hata olu≈ütu: {e}")

    # 2. DataFrame Olu≈ütur ve Aylƒ±k Farklarƒ± Hesapla
    df_raw = pd.DataFrame(raw_list).sort_values('tarih').reset_index(drop=True)
    
    # K√ºm√ºlatiften aylƒ±ƒüa ge√ßi≈ü (Banka verisi fark analizi)
    df_raw['Net_Kar'] = df_raw.groupby('year')['K_Net_Kar'].diff().fillna(df_raw['K_Net_Kar'])
    df_raw['Faiz_Geliri'] = df_raw.groupby('year')['K_Gelir'].diff().fillna(df_raw['K_Gelir'])
    df_raw['Faiz_Gideri'] = df_raw.groupby('year')['K_Gider'].diff().fillna(df_raw['K_Gider'])
    df_raw['Net_Faiz_Marji'] = df_raw.groupby('year')['K_Marj'].diff().fillna(df_raw['K_Marj'])

    # 3. Final Tablo Formatƒ± 
    cols = ['tarih', 'Net_Kar', 'Net_Faiz_Marji', 'Faiz_Geliri', 'Faiz_Gideri']
    df_final = df_raw[cols]
    
    # Kaydet
    df_final.to_excel("Faz2_Bankacilik_Analiz.xlsx", index=False)
    print("\n‚úÖ BA≈ûARILI: 'Faz2_Bankacilik_Analiz.xlsx' dosyasƒ± olu≈üturuldu.")

# Fonksiyonu √áalƒ±≈ütƒ±r
if __name__ == "__main__":
    faz2_olustur()
```
